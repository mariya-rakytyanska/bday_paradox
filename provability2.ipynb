```python
import random 
import matplotlib.pyplot as plt
nb_of_tries = 1000
probabilities = [ ]
```

```python
def has_duplicate (b_list):
    duplicate_present = 0 
    for h in range (len(b_list)):
        for i in range (len(b_list)):
            if b_list[h] == b_list[i] and i != h:
                duplicate_present = 1
                break
        if duplicate_present == 1:
            break
    return duplicate_present  
```

```python
def run_experiment(people, main_list):
    b_list = [ ]
    for x in range (people): 
        x = random.randint(0,365)
        b_list.append(x)
    
    if has_duplicate(b_list) == 1:
        main_list.append(1)
    
```

```python
for student in range (100):
    results_list = [ ]
    for g in range (nb_of_tries):
        run_experiment(student, results_list)
    probabilities.append(len(results_list)/nb_of_tries)
print(probabilities)
```

```python
plt.plot(probabilities)
```
